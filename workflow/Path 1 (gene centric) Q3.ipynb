# Q3: What pathways/processes are [genes] involved in?

## Step 1: Find APIs which connects from Gene to Pathway/Gene Ontology

```python
import requests
import pprint
pp = pprint.PrettyPrinter(indent=2)
apis = requests.get('http://biothings.io/explorer/api/v2/knowledgemap?object.semantic_type=pathway&subject.semantic_type=gene').json()
pp.pprint(apis)
```

```python
apis = requests.get('http://biothings.io/explorer/api/v2/knowledgemap?object.semantic_type=ontology&subject.semantic_type=gene').json()
pp.pprint(apis)
```

```python
results = requests.get('http://biothings.io/explorer/api/v2/directinput2output?input_prefix=ncbigene&output_prefix=reactome.pathway&input_value=1017').json()
pp.pprint(results)
```

```python
results = requests.get('http://biothings.io/explorer/api/v2/directinput2output?input_prefix=ncbigene&output_prefix=go&input_value=1017').json()
pp.pprint(results)
```
