# Q1: What genes are involved in [pathway/process]?

## Step 1: Find APIs which connects from Pathway/Gene Ontology to Gene

```python
import requests
import pprint
pp = pprint.PrettyPrinter(indent=2)
apis = requests.get('http://biothings.io/explorer/api/v2/knowledgemap?object.semantic_type=gene&subject.semantic_type=pathway').json()
pp.pprint(apis)
```

## Step 2: Make API calls to retrieve genes involved in a pathway

```python
results = requests.get('http://biothings.io/explorer/api/v2/directinput2output?input_prefix=reactome.pathway&output_prefix=ncbigene&input_value=R-HSA-187577').json()
pp.pprint(results)
```
